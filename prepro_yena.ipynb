```python
# 날씨데이터 가져오기 
weather = pd.read_csv('weather.csv')
weather.head(30)
```

```python
weather['일시'] = weather['일시'].astype('datetime64[ns]')
weather['월'] = weather['일시'].dt.month
weather['일'] = weather['일시'].dt.day
weather['시간대'] = weather['일시'].dt.hour
```

```python
dlvr = pd.read_csv('delivery.csv')
dlvr.tail()
```

```python
dlvr.info()
```

```python
import datetime

dlvr['일자'] = dlvr['일자'].astype(str)
dlvr['일자'] = pd.to_datetime(dlvr['일자'])
```

```python
dlvr.head()
```

```python
dlvr['월'] = dlvr['일자'].dt.month
dlvr['일'] = dlvr['일자'].dt.day
dlvr.head()
```

```python
df = dlvr.merge(weather, on=['월','일','시간대'])
df.head()
```

```python
df_org = df.drop(['일자','시도','지점','지점명','일시'],axis=1)
df_org
```

```python
df_org['읍면동'].value_counts()
```

```python
yeoksam = df_org[df_org['읍면동']=='역삼동']
yeoksam.tail()
```

```python
# 컬럼 정리해주기
yeoksam = yeoksam.drop(['시군구','읍면동'],axis=1)
yeoksam = yeoksam[['월','일','시간대','요일','기온(°C)','강수량(mm)','풍속(m/s)','풍향(16방위)','습도(%)','적설(cm)','통화건수']]
```

```python
yeoksam.head()
```

```python
yeoksam.columns
```

```python
cols = ['월', '일', '시간대', '요일', '기온(°C)', '강수량(mm)', '풍속(m/s)', '풍향(16방위)','습도(%)', '적설(cm)']

yeoksam = yeoksam.groupby(cols).sum().reset_index()
yeoksam
```
