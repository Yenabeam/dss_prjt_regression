```python
import pandas as pd
```

```python
month_df = pd.DataFrame()
df = pd.DataFrame()
for i in range(1, 13):
    if len(str(i)) == 1:
        mon = "0" + str(i) 
    else:
        mon = str(i)
    month_df = pd.read_csv("./data/CALL_NDELIVERY_{}MONTH.csv".format(mon))
    df = df.append(month_df)
```

```python
df.head()
```

```python
df.sort_values(["일자", "시간대"], inplace=True, ignore_index=True)
df.head()
```

```python
df = df[df["업종"] == "치킨"].reset_index()
```

```python
df = df.drop(columns="업종")
```

```python
df = df.drop(columns="index")
```
